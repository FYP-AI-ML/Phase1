```python
import pandas as pd
import os
from google.colab import files

# Upload the updated.csv file
print("Please upload the updated.csv file.")
uploaded = files.upload()

# Load the uploaded CSV file
csv_filename = next(iter(uploaded.keys()))
df = pd.read_csv(csv_filename)

# Filter the data to get an equal number of real (label=1) and fake (label=0) images
real_images = df[df['label'] == 1]
fake_images = df[df['label'] == 0]

# Shuffle and select 5000 images each
real_images_sample = real_images.sample(n=5000, random_state=1)
fake_images_sample = fake_images.sample(n=5000, random_state=1)

# Combine the samples
selected_images = pd.concat([real_images_sample, fake_images_sample])

# Create a DataFrame for the selected images
selected_images = selected_images.reset_index(drop=True)

# Define the path for the output CSV file
output_csv_path = '/content/base.csv'

# Save the selected images to base.csv
selected_images.to_csv(output_csv_path, index=False)
print(f"Base CSV saved at {output_csv_path}.")

# Provide a download link for the base CSV
files.download(output_csv_path)
```

```python
import pandas as pd
import os
import shutil

# Paths
csv_path = r'D:\FYP\baseFolder\base_updated.csv'  # Path to your CSV file
image_base_folder = r'D:\FYP\baseFolder\crop_img'  # Base path where the images are actually located
destination_folder = r'D:\FYP\splitImage'  # Folder where images will be copied

# Create the destination folder if it doesn't exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Read the CSV file
df = pd.read_csv(csv_path)

# Function to update paths
def update_image_path(old_path):
    # Extract the filename from the old path
    file_name = old_path.split('/')[-1]  # Adjust for forward slashes in old path
    # Construct the new path based on the local directory
    new_path = os.path.join(image_base_folder, file_name)
    return new_path

# Update the paths in the 'path' column
df['updated_path'] = df['path'].apply(update_image_path)

# Now iterate over the updated paths and copy the images
for index, row in df.iterrows():
    image_path = row['updated_path']  # Use the updated path

    # Check if the image exists
    if os.path.isfile(image_path):
        # Copy the image to the destination folder
        shutil.copy(image_path, destination_folder)
        print(f"Copied: {image_path}")
    else:
        print(f"Image not found: {image_path}")

print("All images processed and paths updated.")
```

```python
import pandas as pd
import os

# Paths
csv_path = r'D:\FYP\baseFolder\base_updated.csv'  # Path to the CSV file
split_image_folder = r'D:\FYP\splitImage'  # Path to the splitImage folder

# Read the CSV file
df = pd.read_csv(csv_path)

# Function to check if the image exists in splitImage folder
def check_image_exists(image_path):
    # Extract the filename from the path in CSV
    file_name = os.path.basename(image_path)
    # Construct the full path to the image in the splitImage folder
    split_image_path = os.path.join(split_image_folder, file_name)
    # Check if the image exists
    return os.path.isfile(split_image_path)

# Apply the function to filter rows where the image exists in the splitImage folder
df_filtered = df[df['path'].apply(check_image_exists)]

# Save the updated DataFrame back to CSV
df_filtered.to_csv(csv_path, index=False)

print(f"Updated CSV saved. Rows with missing images have been removed.")
```
